# Processamento da tabela Predict-1


```python
df = pd.read_csv('../../../data/csv/FALHAS_PREDICT-1.csv')
df = pd.DataFrame(df)
df.head()
```

```python
df.columns
```

```python
df = df.drop('Unnamed: 0', axis=1)
df = df.drop(0)
df = df.rename(columns={
    'Unnamed: 1': 'KNR',
    'Unnamed: 2': 'MODELO',
    'Unnamed: 3': 'COR',
    'Unnamed: 4': 'MOTOR',
    'Unnamed: 5': 'ESTACAO',
    'Unnamed: 6': 'USUARIO',
    'Unnamed: 7': 'HALLE',
    'Unnamed: 8': 'FALHA',
    'Unnamed: 9': 'S_GROUP_ID',
    'Unnamed: 10': 'DATA'
})
```

```python
df['MODELO'].value_counts()
```

```python
df.drop(columns='MODELO', inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['FALHA'].value_counts()
```

```python
df['FALHA'] = df['FALHA'].str.upper()
df[['PEÇA', 'FALHA_PEÇA']] = df['FALHA'].str.split(' ', expand=True, n=1)

df['HALLE'] = df['HALLE'].str.split(' ').str[0]
```

```python
df.head(20)
```

```python
count_starts_with_2024 = df['KNR'].str.startswith('2024').sum()
print("Número de KNRs de 2024:", count_starts_with_2024)
count_starts_with_2023 = df['KNR'].str.startswith('2023').sum()
print("Número de KNRs de 2023:", count_starts_with_2023)
```

```python
pivot_df = df.pivot_table(index='KNR', columns='HALLE', aggfunc='size', fill_value=0)
```

```python
pivot_df.columns = [f'QTD_FALHAS_{col}' if col.strip() else 'QTD_FALHAS_SEM_HALLE' for col in pivot_df.columns]
```

```python
pivot_df.reset_index(inplace=True)
```

```python
pivot_df.head()
```

```python
total_zp7 = (df['HALLE'] == 'ZP7').sum()
print(f"Total de ocorrências de ZP7 no df antigo: {total_zp7}")

total_zp7_pivot = (pivot_df['QTD_FALHAS_ZP7']).sum()
print(f"Total de ocorrências de ZP7 no df pivot: {total_zp7_pivot}") 
```

```python
pivot_df.to_parquet('falha_processadas.parquet', index=False)
```

```python
motor_cor_df = df.groupby('KNR')[['MOTOR', 'COR']].first().reset_index()
pivot_df = pd.merge(pivot_df, motor_cor_df, on='KNR', how='left')
```

```python
cols = pivot_df.columns.tolist()

new_order = ['KNR', 'MOTOR', 'COR'] + \
            [col for col in cols if col not in ('KNR', 'MOTOR', 'COR', 'QTD_FALHAS_SEM_HALLE')] + \
            ['QTD_FALHAS_SEM_HALLE']

pivot_df = pivot_df[new_order]
```

```python
pivot_df['MOTOR'] = pivot_df['MOTOR'].replace(r'^\s*$', np.nan, regex=True)
```

```python
pivot_df['TEM_FALHA_ROD'] = np.where(pivot_df['QTD_FALHAS_ROD'] > 0, 1, 0)
```
